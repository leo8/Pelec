## Deux fichiers différents à standardiser et concaténer

On a pas exactement le même nombre de colonnes, ni les mêmes noms de colonnes pour nos deux jeux de données. On trouve des variables dans le jeu de données de 2015 qui ne sont pas présentes dans celui de 2016, et inversement.

Cependant, on veut absolument constituer un dataset commun aux deux années afin de disposer de la quantité maximale de données disponible pour entraîner notre modèle. On va donc mettre à niveau le jeu de données de 2015 en s'alignant sur le schéma du jeu de données de 2016, qui présente une meilleure granularité (notamment sur les adresses). Pour le dataset de 2015, on va supprimer les colonnes 2010 Census Tracts, Seattle Police Department, Micro Community Policing Plan Areas, City Council Districts, OtherFuelUse(kBtu) et SPD Beats. Puis, on va supprimer la colonne Zip Codes, et extraire de la colonne Location les variables 'Latitude, Longitude, Address, City, State. On va également renommer les colonnes GHGEmissions(MetricTonsCO2e), GHGEmissionsIntensity(kgCO2e/ft2) et Comment par leurs noms respectifs dans le jeu de données de 2016, soit TotalGHGEmissions, GHGEmissionsIntensity et Comments.

```python
raw_data_2015 = pd.read_csv('2015-building-energy-benchmarking.csv')
raw_data_2016 = pd.read_csv('2016-building-energy-benchmarking.csv')
```

```python
raw_data_2015.drop(['Zip Codes',
                    '2010 Census Tracts',
                    'Seattle Police Department Micro Community Policing Plan Areas',
                    'City Council Districts',
                    'OtherFuelUse(kBtu)',
                    'SPD Beats'],
                   axis=1, inplace=True)
```

<font color="#337da4"><h3>3. La question des doublons</h3><a class="anchor" id="section_3"></a></font>

## Traiter les doublons générés par la concaténation

Du fait de l'étape précédente de concaténation, on obtient des doublons des bâtiments dans notre dataset. Il faut traiter ces doublons en faisant la distinction entre deux cas :

- Si les caractéristiques du bâtiment ont changé entre 2015 et 2016, on peut conserver le doublon, en considérant qu'il s'agit d'un nouveau bâtiment : en effet, il s'agit de nouvelles données intéressantes pour entraîner notre modèle. Il faudra cependant veiller à modifier l'ID du building pour faire la distinction entre les deux bâtiments.
- Au contraire, si les caractéristiques du bâtiment n'ont pas changé d'une année à l'autre, il faut supprimer un des deux doublons. A ce moment là, on va modifier les informations relatives à la consommation d'énergie, en les remplaçant par la moyenne des deux années.

```python
BUILDING_PROPERTY_FEATURES = ['OSEBuildingID','PrimaryPropertyType', 'BuildingType',
       'CouncilDistrictCode',
       'Neighborhood', 'YearBuilt', 'NumberofBuildings', 'NumberofFloors',
       'PropertyGFATotal', 'PropertyGFAParking', 'PropertyGFABuilding(s)']

ENERGY_FEATURES = ["Electricity(kWh)", "TotalGHGEmissions"]

l_drop = []

# Comptage des batiments par occurence
from collections import Counter
building_id_list = raw_data["OSEBuildingID"].tolist()
counter = Counter(building_id_list)

# Copie et normalisation du dataframe initial
data = raw_data.copy()
data = data.replace('\n','', regex=True)
data = data.applymap(lambda s: s.lower() if type(s) == str else s)

for building_id in counter.keys():
    if counter[building_id] > 1: # On traite les batiments en double
        # Dataframe tavaec les deux lignes dupliquées
        tmp_df = data[BUILDING_PROPERTY_FEATURES][data["OSEBuildingID"]==building_id]
        tmp_full_df = data[data["OSEBuildingID"]==building_id]
       
        # On récupère l'index des deux lignes dans le dataframe initial
        first_index = tmp_df.index[0]
        second_index = tmp_df.index[1]
       
        # Si les 2 lignes ne sont pas similaires, on renomme les OSEBuildingID
        if False in (tmp_df.iloc[0] == tmp_df.iloc[1]).tolist():
            data.loc[first_index, "OSEBuildingID"] = str(data.loc[first_index, "OSEBuildingID"]) + "_2015"
            data.loc[second_index, "OSEBuildingID"] = str(data.loc[second_index, "OSEBuildingID"]) + "_2016"
           
        # Si les 2 lignes sont similaires, on fusionne par la moyenne
        else:
            l_drop.append(second_index)
            for energy_feature in ENERGY_FEATURES:
                data.loc[first_index, energy_feature] = tmp_full_df[energy_feature].mean()
                
data = data.drop(l_drop)
```

Analysons les données uniques des colonnes.

<b>Explication des variables</b><br>

Plusieurs variables concernent les surfaces (GFA = Gross floor area) :<br>
- Gross floor area : surface totale
- PropertyGFAParking : surface des parkings
- PropertyGFABuilding(s) : surface des bâtiments
- ListOfAllPropertyUseTypes : les différentes utilisations des surfaces (Hotel, Parking, Restaurant, Piscine, etc.)
- LargestPropertyUseType / LargestPropertyUseTypeGFA : principale destination du bâtiment et sa surface
- SecondLargestPropertyUseType / SecondLargestPropertyUseTypeGFA : seconde destination du bâtiment et sa surface
- ThirdLargestPropertyUseType	/ ThirdLargestPropertyUseTypeGFA : troisième destination du bâtiment et sa surface

Concernant l'énergie, ENERGY STAR nous indique :
- SiteEUI(kBtu/sf)	/ SiteEUIWN(kBtu/sf)	: Energy Use Intensity (EUI) ou intensité de la consommation d'énergie ; l'EUI exprime la consommation d'énergie d'un bâtiment en fonction de sa taille ou d'autres caractéristiques.<br>
- SourceEUI(kBtu/sf)	SourceEUIWN(kBtu/sf) : la source d'énergie est l'unité d'évaluation la plus équitable pour comparer différents bâtiments entre eux. La source d'énergie représente la quantité totale de combustible brut nécessaire au fonctionnement du bâtiment. Il intègre toutes les pertes de transmission, de livraison et de production. En prenant en compte toutes les consommations d'énergie, le score fournit une évaluation complète de l'efficacité énergétique d'un bâtiment. <br>

SiteEnergyUse(kBtu)	SiteEnergyUseWN(kBtu)	SteamUse(kBtu)	Electricity(kWh)	Electricity(kBtu)	NaturalGas(therms)	NaturalGas(kBtu) : l'utilisation des énergies est exprimée en diverses unités :
- kBTu : le B.T.U. (British Thermal Unit) est une unité de mesure d'énergie thermique. Le BTU, de symbole Btu, est utilisé dans les pays anglo-saxons et asiatiques et équivaut à 1055,06 Joule ou 252 calories.<br>
1 BTU est l’énergie nécessaire pour chauffer ou refroidir une livre d’eau de 1 degré Fahrenheit au niveau de la mer.<br>
Pour convertir une puissance en Btu une puissance en watt il suffit de diviser la puissance en Btu par 3414,5 pour obtenir la puissance en Kw. <br>

- therms : le therm est une unité d'énergie thermique anglo-saxone égale à 1 000 000 unités thermiques britanniques (BTU). 
>1 kWh = 0.034095106405145 thm <br>
>1 thm = 29.329722222222 kWh

Supprimons les colonnes qui ne nous intéressent pas dans ce projet :

D'autres variables sont redondantes :

- Electricity(kWh) / Electricity(kBtu)
- NaturalGas(therms) / NaturalGas(kBtu)
- TotalGHGEmissions / GHGEEmissionsIntensity

Enfin, pour ce qui est des variables relatives à l'énergie, nous pouvons supprimer les données surfaçiques, qui sont là encore, d'une certaine manière, redondantes : 

- SiteEUI(kBtu/sf)
- SourceEUI(kBtu/sf)

```python
def search_componant(df, suffix=None):
  componant = []
  for col in df.columns:
      if suffix in col: 
        componant.append(col)
  return componant
```

Il peut être intéressant de traiter l'âge des bâtiments, plutôt que leur année de construction, pour réduire la dispersion des données et lier l'année des relevés. Nous allons donc créer une nouvelle variable BuildingAge et supprimer l'année de construction.

```python
data['BuildingAge'] = data['DataYear'] - data['YearBuilt']
data.drop('YearBuilt', axis=1, inplace=True)
```

```python
fig = plt.figure(figsize=(10,7))
ax = sns.histplot(data=data, x='BuildingAge', bins=int((data.BuildingAge.max() - data.BuildingAge.min())/5))
ax.set_xlabel("Age du bâtiment")
ax.set_ylabel("Nombre de bâtiments")
plt.title(f"Distribution de l'âge des bâtiments\n", fontdict=font_title)
plt.show()
```

<h4><font color="#337da4">3.3 Corrélations</font></h4>

```python
#This function allows you to view the distribution 
#of CO2 emissions within the classes of a variable.
def visualisation(variable, df, ylim):
    # the_mean=df["GHGEmissions(MetricTonsCO2e)"].mean()
    the_mean=df["TotalGHGEmissions"].mean()
    fig=plt.figure(figsize=[18,7])
    fig.patch.set_facecolor('#E0E0E0')
    fig.patch.set_alpha(0.7)
    plt.title("C02 emissions distribution by {}".format(variable),size=16)
    sns.boxplot(x=variable, y="TotalGHGEmissions", data=df,color="#cbd1db",width=0.5,showfliers=False,showmeans=True)
    plt.hlines(y=the_mean,xmin=-0.5,xmax=len(df[variable].unique())-0.5,color="#6d788b",ls="--",label="Global mean")

    plt.ylabel(" C02 emissions",size=14)
    plt.xticks(range(0,len(df[variable].unique()))
               ,df[variable].unique(),rotation=90)
    
    plt.ylim(ylim)
    plt.legend()
    plt.grid()
    plt.show()
```

```python
visualisation('Neighborhood',data, ylim=[0,600])
```

```python
visualisation('BuildingType',data, ylim=[0,350])
```

```python
corr = data.corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
fig, ax = plt.subplots(figsize=(10,10))
ax = sns.heatmap(corr, annot=True, fmt=".2f", annot_kws={'size':8}, 
                 mask=mask, center=0, cmap="coolwarm")
plt.title(f"Heatmap des corrélations linéaires\n", fontsize = 18)
plt.show()
```

On peut observer des corrélations importantes entre les variables :

- PropertyGFATotal d'une part, et PropertyGFABuilding(s) / LargestPropertyUseTypeGFA de l'autre;
- PropertyGBABuilding(s) et LargestPropertyUseTypeGFA;
- TotalGHEmissions et SiteEnergyUse(kBtu) / NaturalGas(kBtu);

Pour les variables à prédire 'TotalGHGEmissions' et 'SiteEnergyUse(kBtu)', on remarque également de fortes corrélations avec les variables de relevés (les consommations), le nombre de batiments ou d'étages, et les surfaces au sol.
